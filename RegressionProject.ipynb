# Multilinear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

X_train = df_train[['SQUARE_FT', 'LONGITUDE', 'LATITUDE']]
y_train = df_train['TARGET(PRICE_IN_LACS)']

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)

mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print("Training Performance:")
print(f"Mean Absolute Error: {mae_train}")
print(f"Mean Squared Error: {mse_train}")
print(f"R-squared: {r2_train}")

plt.scatter(y_train, y_train_pred, alpha=0.3)
plt.xlabel('Actual Prices (in lacs)')
plt.ylabel('Predicted Prices (in lacs)')
plt.title('Actual vs Predicted Prices (Training Set)')
plt.show()

X_test = df_test[['SQUARE_FT', 'LONGITUDE', 'LATITUDE']]
y_test_pred = model.predict(X_test)

df_test['PREDICTED_PRICE_IN_LACS'] = y_test_pred

df_test.to_csv('test_predictions.csv', index=False)

print("Predictions saved to 'test_predictions.csv'")
```

# REducing Overfitting in Decision Tree Regression

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

feature_columns = ['SQUARE_FT', 'LONGITUDE', 'LATITUDE']

X_train = train_df[feature_columns]
y_train = train_df['TARGET(PRICE_IN_LACS)']

X_test = test_df[feature_columns]

if 'ID' not in test_df.columns:
    test_df['ID'] = range(1, len(test_df) + 1)

X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

dt_regressor = DecisionTreeRegressor(max_depth=5, min_samples_split=10, min_samples_leaf=5, max_leaf_nodes=20)
dt_regressor.fit(X_train_split, y_train_split)

y_val_pred = dt_regressor.predict(X_val)

mse_val = mean_squared_error(y_val, y_val_pred)
r2_val = r2_score(y_val, y_val_pred)

print(f'Mean Squared Error on validation set: {mse_val}')
print(f'R-squared on validation set: {r2_val}')

y_train_pred = dt_regressor.predict(X_train_split)
mse_train = mean_squared_error(y_train_split, y_train_pred)
r2_train = r2_score(y_train_split, y_train_pred)

print(f'Mean Squared Error on training set: {mse_train}')
print(f'R-squared on training set: {r2_train}')

y_test_pred = dt_regressor.predict(X_test)

submission = pd.DataFrame({'ID': test_df['ID'], 'Predicted Price': y_test_pred})
submission.to_csv('submission.csv', index=False)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

feature_columns = ['SQUARE_FT', 'LONGITUDE', 'LATITUDE']

X_train = train_df[feature_columns]
y_train = train_df['TARGET(PRICE_IN_LACS)']

X_test = test_df[feature_columns]

X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

dt_regressor = DecisionTreeRegressor(max_depth=10, min_samples_split=5, min_samples_leaf=2)
dt_regressor.fit(X_train_split, y_train_split)

y_val_pred = dt_regressor.predict(X_val)

mse_val = mean_squared_error(y_val, y_val_pred)
r2_val = r2_score(y_val, y_val_pred)

print(f'Mean Squared Error on validation set: {mse_val}')
print(f'R-squared on validation set: {r2_val}')

y_train_pred = dt_regressor.predict(X_train_split)
mse_train = mean_squared_error(y_train_split, y_train_pred)
r2_train = r2_score(y_train_split, y_train_pred)

print(f'Mean Squared Error on training set: {mse_train}')
print(f'R-squared on training set: {r2_train}')

print("\nEnter the details of the house to predict its price:")
square_ft = float(input("Enter SQUARE_FT: "))
longitude = float(input("Enter LONGITUDE: "))
latitude = float(input("Enter LATITUDE: "))

user_data = pd.DataFrame({
    'SQUARE_FT': [square_ft],
    'LONGITUDE': [longitude],
    'LATITUDE': [latitude]
})

predicted_price = dt_regressor.predict(user_data)[0]
print(f"\nPredicted Price: {predicted_price:.2f} Lacs")
```
